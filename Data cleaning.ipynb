```python
import pandas as pd

Healthcare_data = pd.read_excel('dirty_healthcare-data.xlsx')


Healthcare_data.head()
```

```python
Healthcare_data.info()  
Healthcare_data.describe()  
Healthcare_data.isnull().sum()  
```

```python
Healthcare_data[Healthcare_data.duplicated()] 
```

```python
num_duplicates = Healthcare_data.duplicated().sum()
num_duplicates
```

```python
Healthcare_data[Healthcare_data.duplicated(subset=['ID'])]
```

```python
Healthcare_data = Healthcare_data.drop_duplicates(subset=['ID'], keep='first') 
```

```python
Healthcare_data
```

```python
Healthcare_data['Name'] = Healthcare_data ['Name'].str.title()
Healthcare_data ['City'] = Healthcare_data ['City'].str.title()

```

```python
Healthcare_data 
```

```python
Healthcare_data['Blood Type'] = Healthcare_data['Blood Type'].fillna('Unknown')
Healthcare_data['Education'] = Healthcare_data['Education'].fillna('Unknown')
Healthcare_data['Health Condition'] = Healthcare_data['Health Condition'].fillna('Unknown')
Healthcare_data['Gender'] = Healthcare_data['Gender'].fillna('Unknown')
```

```python
Healthcare_data['Gender'] = Healthcare_data['Gender'].replace('M', 'Male')
Healthcare_data['Gender'] = Healthcare_data['Gender'].replace('F', 'Female')
print(Healthcare_data['Gender'].unique())
```

```python
Healthcare_data
```

```python
Healthcare_data['Salary'] = Healthcare_data['Salary'].replace(
    {r'\$': '', r'\(': '', r'\)': '', 'unstable': '', 'Commission': '', r',': '', 'Seasonal': ''}, regex=True)
Healthcare_data
```

```python
Healthcare_data['Salary'] = pd.to_numeric(Healthcare_data['Salary'], errors='coerce')

Healthcare_data['Salary'] = Healthcare_data['Salary'].apply(lambda x: abs(x) if pd.notnull(x) and x < 0 else x)

Healthcare_data['Salary'] = Healthcare_data['Salary'].fillna(0)

Healthcare_data
```

```python
Healthcare_data['Age'] = pd.to_numeric(Healthcare_data['Age'], errors='coerce')

Healthcare_data['Age'] = Healthcare_data['Age'].fillna(0).astype(int)

Healthcare_data.head()
```

```python
Healthcare_data['Credit Score'] = pd.to_numeric(Healthcare_data['Credit Score'], errors='coerce')
Healthcare_data['Credit Score'] = Healthcare_data['Credit Score'].fillna(0).astype(int)
Healthcare_data['Credit Score'] = Healthcare_data['Credit Score'].fillna(0)
Healthcare_data
```

```python
Healthcare_data['Date of Admission'] = pd.to_datetime(Healthcare_data['Date of Admission'], errors='coerce')
Healthcare_data
```
